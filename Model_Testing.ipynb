```python
def initialize_random_strategy_matrix(n, enemies, powers):
    X = np.zeros((n, n))
    for i in range(n):
        # Random allocations summing to p_i
        allocations = np.random.dirichlet(np.ones(len(enemies[i]) + 1)) * powers[i]
        X[i][i] = allocations[0]
        for idx, j in enumerate(enemies[i]):
            X[i][j] = allocations[idx + 1]
            X[j][i] = X[i][j]  # Ensure symmetry
    return X
```

```python
def compute_state_i(X, i):
    xii = X[i][i]
    total_attack_on_i = sum(X[j][i] for j in range(len(X)) if j != i)
    if xii > total_attack_on_i:
        return 'safe'
    elif xii == total_attack_on_i:
        return 'precarious'
    else:
        return 'unsafe'

def utility_i(X, i, allies, enemies):
    state_i = compute_state_i(X, i)
    if state_i == 'unsafe':
        return 0

    utility = 0
    for j in allies[i] + enemies[i]:
        state_j = compute_state_i(X, j)
        if j in allies[i]:
            if state_j == 'unsafe':
                uij = 0
            else:
                uij = 1
        elif j in enemies[i]:
            if state_j == 'safe':
                uij = 0
            else:
                uij = 1
        utility += uij
    return utility
```

```python
from scipy.optimize import minimize

def preferred_response(X, i, pi, allies, enemies, type):
    n = X.shape[0]
    indices = [i] + enemies[i]  # Decision variables: x_{ii} and x_{ij} for enemies

    # Initial guess for decision variables
    x0 = [X[i][j] for j in indices]

    # Bounds for variables (0 to pi)
    bounds = [(0, pi) for _ in indices]

    # Constraint: Sum of allocations equals pi
    constraints = {
        'type': 'eq',
        'fun': lambda x_i_values: np.sum(x_i_values) - pi
    }

    def objective(x_i_values):
        # Update X_copy with x_i_values
        X_copy = X.copy()
        for idx, j in enumerate(indices):
            X_copy[i][j] = x_i_values[idx]
            X_copy[j][i] = x_i_values[idx]  # Symmetry
        # Negative utility because we are minimizing
        if type == 0:
            return -utility_i(X_copy, i, allies, enemies)
        elif type == 1:
            return -utility_econ_i(X_copy, i, pi, allies, enemies, alpha[i])

    # Optimize
    result = minimize(objective, x0, bounds=bounds, constraints=constraints)

    # Update X if optimization is successful
    if result.success:
        x_i_best = result.x
        for idx, j in enumerate(indices):
            X[i][j] = x_i_best[idx]
            X[j][i] = x_i_best[idx]
    else:
        print(f"Optimization failed for country {i}")

    return X
```

## Strategy + Economic Stability Model

```python
def utility_econ_i(X, i, p_i, allies, enemies, alpha_i):
    state_i = compute_state_i(X, i)
    if state_i == 'unsafe':
        return 0  # No utility if country is unsafe

    utility = 0
    allocated_power = X[i][i]  # Power allocated to self-defense

    for j in allies[i] + enemies[i]:
        state_j = compute_state_i(X, j)
        if j in allies[i]:
            if state_j == 'unsafe':
                uij = 0
            else:
                uij = 1
        elif j in enemies[i]:
            if state_j == 'safe':
                uij = 0
            else:
                uij = 1
        utility += uij
        allocated_power += X[i][j]

    # Calculate allocation to economic stability
    allocation_to_economy = p_i - allocated_power
    allocation_to_economy = max(allocation_to_economy, 0)  # Ensure non-negative

    # Add economic stability component to utility
    utility += alpha_i * allocation_to_economy

    return utility
```

```python
N = 100
nash_set_econ = set()

for sim in range(N):
    X = initialize_random_strategy_matrix(n, enemies, pi)
    max_iterations = 1000
    tolerance = 1e-6

    for iteration in range(max_iterations):
        X_prev = X.copy()
        for i in range(n):
            X = preferred_response(X, i, pi[i], allies, enemies, 1)
        diff = np.linalg.norm(X - X_prev)
        if diff < tolerance:
            nash_set_econ.add(tuple(X.flatten()))
            print(f"Simulation {sim + 1}: Converged to Nash equilibrium after {iteration + 1} iterations.")
            break
    else:
        print(f"Simulation {sim + 1}: Did not converge within the maximum number of iterations.")
```

```python
nash_list_econ = list(nash_set_econ)

amount_survive_econ = {}

for index in range(n_countries):
    amount_survive_econ[index] = 0

state_to_value = {
        'safe' : 0,
        'precarious' : 1,
        'unsafe' : 2
}

for nash in nash_list_econ:
    strat_matrix = np.array(nash).reshape(n, n)
    for index in amount_survive_econ.keys():
        if state_to_value[compute_state_i(strat_matrix, index)] < 2:
            amount_survive_econ[index] += 1
```

```python
prob_survive_econ = {}

for index in amount_survive_econ.keys():
    prob_survive_econ[index] = amount_survive_econ[index]/len(nash_list_econ)
```
