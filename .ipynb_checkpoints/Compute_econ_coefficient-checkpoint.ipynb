```python
regions = {"north_africa" : north_africa, 
           "west_africa" : west_africa, 
           "central_africa" : central_africa, 
           "east_africa" : east_africa, 
           "southern_africa" : southern_africa,
           "middle_east" : middle_east, 
           "central_asia" : central_asia, 
           "south_asia" : south_asia, 
           "southeast_asia" : southeast_asia, 
           "east_asia" : east_asia, 
           "eastern_europe" : eastern_europe,
           "northern_europe" : northern_europe, 
           "southern_europe" : southern_europe, 
           "western_europe" : western_europe, 
           "central_america" : central_america, 
           "caribbean" : caribbean,
           "south_america" : south_america, 
           "north_america" : north_america,
           "oceania" : oceania}
```

```python
def classify_region(x):
    for region in regions.keys():
        if x in regions[region]:
            return region
```

```python
gdp_g_89['Region'] = gdp_g_89['Country Name'].apply(lambda x: classify_region(x))
ir_89['Region'] = ir_89['Country Name'].apply(lambda x: classify_region(x))
gdp_gpc_89['Region'] = gdp_gpc_89['Country Name'].apply(lambda x: classify_region(x))
gcf_89['Region'] = gcf_89['Country Name'].apply(lambda x: classify_region(x))
igs_89['Region'] = igs_89['Country Name'].apply(lambda x: classify_region(x))
```

```python
for dataset in datasets:
    for region in regions.keys():
        print(region)
        print(dataset[dataset['Region'] == region]['1989'].isna().sum()/len(dataset[dataset['Region'] == region]['1989']))
```

```python
def impute_data(data):
    region_median = data.groupby('Region')['1989'].median().to_dict()
    for region in regions.keys():
        data.loc[data['Region'] == region, '1989'] = data.loc[data['Region'] == region, '1989'].fillna(region_median[region])
    return data
```

```python
gdp_g_89 = impute_data(gdp_g_89)
ir_89 = impute_data(ir_89)
gdp_gpc_89 = impute_data(gdp_gpc_89)
gcf_89 = impute_data(gcf_89)
igs_89 = impute_data(igs_89)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Example of normalization
scaler = MinMaxScaler()
economic_features = econ_data[['gdp_g', 'ir', 'gdp_gpc', 'gcf', 'igs']]
normalized_features = scaler.fit_transform(economic_features)
```

```python
econ_data['alpha'] = (0.3 * normalized_features[:, 0] +  # GDP growth
                 0.2 * normalized_features[:, 1] +  # GDP per capita growth
                 0.2 * normalized_features[:, 2] -  # Gross capital formation
                 0.2 * normalized_features[:, 4] -  # Inflation rate
                 0.1 * normalized_features[:, 3])   # Imports of goods and services
```
